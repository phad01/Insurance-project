```python
# Replace the entity, Joint gender, sex and not stated value in gender with NaN value
V_insurance["Gender"] = V_insurance["Gender"].replace('Entity', np.nan)
V_insurance["Gender"] = V_insurance["Gender"].replace('Joint Gender', np.nan)
V_insurance["Gender"] = V_insurance["Gender"].replace('SEX', np.nan)
V_insurance["Gender"] = V_insurance["Gender"].replace('NO GENDER', np.nan)
V_insurance["Gender"] = V_insurance["Gender"].replace('NOT STATED', np.nan)
V_insurance["Gender"].unique()
```

```python
# fill NaNs in all columns with the mode of the all columns
# Reconfirm that all columns has no NaNs again.
V_insurance['Gender'] = V_insurance['Gender'].fillna(V_insurance['Gender'].mode()[0])  
V_insurance['Car_Category'] = V_insurance['Car_Category'].fillna(V_insurance['Car_Category'].mode()[0])
V_insurance['Subject_Car_Colour'] = V_insurance['Subject_Car_Colour'].fillna(V_insurance['Subject_Car_Colour'].mode()[0])
V_insurance['Subject_Car_Make'] = V_insurance['Subject_Car_Make'].fillna(V_insurance['Subject_Car_Make'].mode()[0])
V_insurance['LGA_Name'] = V_insurance['LGA_Name'].fillna(V_insurance['LGA_Name'].mode()[0])
V_insurance['State'] = V_insurance['State'].fillna(V_insurance['State'].mode()[0])
V_insurance.info()
```

```python
# Changing the Policy start date, End date and first transaction date to datetime data-type

V_insurance['Policy Start Date'] =  pd.to_datetime(V_insurance['Policy Start Date'])
V_insurance['Policy End Date'] =  pd.to_datetime(V_insurance['Policy End Date'])
V_insurance['First Transaction Date'] = pd.to_datetime(V_insurance['First Transaction Date'])
```

```python
# Extracting the difference between both the policy start and end date
V_insurance['Policy Period'] = abs(V_insurance ['Policy End Date'] - V_insurance['Policy Start Date'])
```

```python
# Converting the Policy Period to int type.
V_insurance['Policy Period'] = V_insurance['Policy Period'].dt.days.astype('int64')
```

```python
V_insurance.info()
```

```python
# Reordering the columns to ensure the outcome is on the last column
column_names = ["ID","Policy Start Date","Policy End Date","Policy Period","Gender","Age","First Transaction Date", "No_Pol","Car_Category","Subject_Car_Colour","Subject_Car_Make","LGA_Name","State", "ProductName", "target"]
V_insurance = V_insurance.reindex(columns = column_names)
```
